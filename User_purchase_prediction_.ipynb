In pandas DataFrames, iloc stands for integer-location based indexing. It's a method used to select specific rows and columns by their numerical positions within the DataFrame, regardless of any custom labels or names assigned to those rows or columns.

X=df.iloc[:,0:2]

Selects columns:
: in the first position ([:,...]) indicates selecting all rows (from the beginning : to the end :).
0:2 specifies the column range to include: from the 0th index (column A) up to, but not including, the 2nd index (column B). So, it selects columns A and B.
Assigns the result (a DataFrame containing columns A and B) to the variable X.



y=df.iloc[:,-1]

Selects a single column:
: in the second position ([...,:]) indicates selecting all columns (from the beginning : to the end :).
-1 refers to the last column (C in this example) using zero-based indexing.
Assigns the result (a pandas Series containing the values from column C) to the variable y.
In simpler terms:

X becomes a new DataFrame containing only the first two columns (A and B) from the original df.
y becomes a Series containing the values from the last column (C) of df.

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
X_scaled.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model= Sequential()

model.add(Dense(10,activation='relu',input_dim=2))                #input_dim means input dimension
model.add(Dense(10,activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
print("X_scaled shape:", X_scaled.shape)
print("y shape:", y.shape)
```

```python
print("Model output shape:", model.output_shape)
```

```python
#compiling the model

model.compile(loss='binary_crossentropy', metrics=['accuracy'])
```

batch size=1, it means fir the training we give 1 row for the trainig (number to datapoint go for training at 1 time.) and if the batch_size=320 means we are given 320 rows for the process.

```python
# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=1)  #vanila Gradient decent
```
